```python
import pandas as pd
import requests
```

```python
Category = ['NIFTY','BANKNIFTY','SME','FO','OTHERS','ALL']
```

```python
import requests

# Set up the headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36",
    "Accept": "application/json, text/plain, */*",
    "Referer": "https://www.nseindia.com",
    "Accept-Language": "en-US,en;q=0.9",
}
all_data = []
for x in Category:

  # URL for the API
  url = "https://www.nseindia.com/api/market-data-pre-open?key=" + x

  # Send the request
  try:
      session = requests.Session()
      # Make an initial request to obtain necessary cookies
      session.get("https://www.nseindia.com", headers=headers)
      response = session.get(url, headers=headers)

      # Check if the request was successful
      if response.status_code == 200:
          data = response.json()
          print(data)  # Print the data or process it as needed
          all_data.append(data)
      else:
          print(f"Failed to fetch data. Status Code: {response.status_code}")
  except Exception as e:
      print(f"An error occurred: {e}")
```

```python
all_data_dict = {'Sector':[],
                 'Stock Name':[],
                 'Buy':[],
                 'Sell':[]}
```

```python
Category = ['NIFTY 50','Nifty Bank','Emerge','Securities in F&O','Others','All']
```

```python
for x in range(len(all_data)):
  sector = Category[x]
  data = all_data[x]['data']
  for y in range(len(data)):
    all_data_dict['Sector'].append(sector)
    all_data_dict['Stock Name'].append(data[y]['metadata']['symbol'])
    ato = data[y]['detail']['preOpenMarket']['ato']
    all_data_dict['Buy'].append(ato['totalBuyQuantity'])
    all_data_dict['Sell'].append(ato['totalSellQuantity'])
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(all_data_dict)
df
```
